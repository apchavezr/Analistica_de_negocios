# Visualización de KPIs para el sector retail (simulado)

```python
# 1. Importación de librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# 2. Carga de datos
df = pd.read_csv('datos_kpis_retail.csv')  # Asegúrese de tener el archivo en el mismo directorio
```

### Gráfico 4: Ventas Totales por Mes

Este gráfico de línea muestra la evolución de las ventas totales mes a mes. Permite identificar tendencias, estacionalidades o caídas en el ingreso mensual de la tienda retail. Un aumento sostenido sugiere crecimiento, mientras que descensos abruptos pueden indicar eventos adversos o problemas operativos.


```python
# 3. Vista previa de los datos
print(df.head())
```

### Gráfico 5: Clientes Activos por Mes

El gráfico de barras refleja la cantidad de clientes únicos que interactuaron con el negocio en cada mes. Esta métrica ayuda a comprender el alcance de la tienda y evaluar la efectividad de las campañas de captación de nuevos usuarios o la fidelización de clientes existentes.


```python
# 4. Gráfico: Ventas Totales por Mes
plt.figure()
sns.lineplot(data=df, x='Mes', y='Ventas_Totales_USD', marker='o')
plt.title('Ventas Totales por Mes (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Gráfico 6: Tasa de Conversión por Mes

Aquí se analiza qué porcentaje de visitantes terminó comprando. Una tasa de conversión estable y creciente es señal de un embudo de ventas eficiente. Si el valor es bajo, puede ser necesario revisar la experiencia de usuario, los precios o la oferta de productos.


```python
# 5. Gráfico: Clientes Activos por Mes
plt.figure()
sns.barplot(data=df, x='Mes', y='Clientes_Activos', palette='Blues_d')
plt.title('Clientes Activos por Mes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Gráfico 7: Carritos Abandonados

Este gráfico ayuda a identificar la proporción de oportunidades de venta que no se concretaron. Una cifra alta puede estar relacionada con problemas en el proceso de compra, costos de envío, tiempos de entrega o desconfianza del cliente.


```python
# 6. Gráfico: Tasa de Conversión por Mes
plt.figure()
sns.lineplot(data=df, x='Mes', y='Tasa_Conversion', marker='o', color='green')
plt.title('Tasa de Conversión Mensual')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Gráfico 8: Valor Promedio por Pedido

Muestra cuánto gasta en promedio un cliente cada vez que realiza una compra. Este KPI es útil para planear estrategias de *upselling* y *cross-selling*. Un valor bajo podría indicar compras impulsivas de bajo costo, mientras que uno alto podría reflejar lealtad o éxito en promociones.


```python
# 7. Gráfico: Carritos Abandonados
plt.figure()
sns.barplot(data=df, x='Mes', y='Carritos_Abandonados', color='salmon')
plt.title('Carritos Abandonados por Mes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 8. Gráfico: Valor Promedio por Pedido
plt.figure()
sns.lineplot(data=df, x='Mes', y='Valor_Promedio_Pedido', marker='o', color='purple')
plt.title('Valor Promedio por Pedido (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

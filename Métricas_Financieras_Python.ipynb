# Cálculo de métricas financieras con datos simulados

Este notebook muestra cómo calcular tres métricas financieras clave: **Margen Bruto**, **Retorno sobre la Inversión (ROI)** y **Customer Lifetime Value (CLTV)** usando datos simulados.

```python
import pandas as pd
import numpy as np

# Simulación de datos
np.random.seed(0)
meses = pd.date_range(start='2024-01-01', periods=12, freq='M').strftime('%Y-%m')
ingresos = np.random.randint(50000, 100000, size=12)
costos = np.random.randint(20000, 50000, size=12)
inversiones = np.random.randint(10000, 30000, size=12)
ganancia_neta = ingresos - costos

# Datos para CLTV
valor_promedio_pedido = np.random.uniform(50, 200, size=12)
frecuencia_compra = np.random.randint(1, 5, size=12)
tiempo_vida_cliente = np.random.randint(6, 24, size=12)

# Crear DataFrame
df = pd.DataFrame({
    'Mes': meses,
    'Ingresos': ingresos,
    'Costos': costos,
    'Ganancia_Neta': ganancia_neta,
    'Inversiones': inversiones,
    'Margen_Bruto_%': ((ingresos - costos) / ingresos * 100).round(2),
    'ROI_%': (ganancia_neta / inversiones * 100).round(2),
    'CLTV': (valor_promedio_pedido * frecuencia_compra * tiempo_vida_cliente).round(2)
})

df.head()
```

## Visualización de las métricas financieras

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df['Mes'], df['Margen_Bruto_%'], marker='o', label='Margen Bruto (%)')
plt.plot(df['Mes'], df['ROI_%'], marker='s', label='ROI (%)')
plt.xticks(rotation=45)
plt.title('Margen Bruto y ROI por Mes')
plt.ylabel('%')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Mes'], df['CLTV'], color='purple')
plt.xticks(rotation=45)
plt.title('Customer Lifetime Value (CLTV) por Mes')
plt.ylabel('CLTV (USD)')
plt.tight_layout()
plt.show()
```

## Conclusiones

- El **Margen Bruto** permite observar qué tan eficiente es la empresa al convertir ventas en ganancias antes de otros gastos.
- El **ROI** refleja qué tan rentables fueron las inversiones realizadas cada mes.
- El **CLTV** estimado da una idea del valor que cada cliente puede generar, útil para estrategias de retención y segmentación.
